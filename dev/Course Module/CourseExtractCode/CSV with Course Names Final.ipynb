```python
def parse1(soup):
    spec=[];
    for ul in soup.select('ul'):
        for li in ul.select('li'):
            temp=li.text
            a=hasNumbers(temp)
            if thing1 and a:
                spec.append(li.text)  
    return spec

def parse2(soup):
    spec=[];
    for ul in soup.select('ul'):
        for li in ul.select('li'):
            temp=li.text
            a=hasNumbers(temp)
            if thing1 and a:
                spec.append(li.text)      
    return spec

def parse3(soup):
    spec=[];
    for ul in soup.select('tbody'):
        for li in ul.select('tr'):
            temp=li.text
            a=hasNumbers(temp)
            if thing1 and a:
                spec.append(li.text)      
    return spec

def parse4(soup):
    spec=[];
    for ul in soup.select('p'):
        for li in ul.select('strong'):
                spec.append(li.text)      
    return spec

def parse5(soup):
    spec=[];
    for ul in soup.select('li'):
        for li in ul.select('h3'):
            spec.append(li.text) 
    return spec

def parse6(soup):
    spec=[];
    for ul in soup.select('li'):
        for li in ul.select('strong'):
                spec.append(li.text)      
    return spec

def parse7(soup):
    spec=[];
    for ul in soup.select('li'):
        for li in ul.select('span'):
            a=hasNumbers(li.text)
            if thing1 and a:
                spec.append(li.text)
    return spec

def parse8(soup):
    spec=[];
    for ul in soup.select('div'):
        for li in ul.select('p'):
            a=hasNumbers(li.text)
            if thing1 and a:
                spec.append(li.text)
    return spec


def parse9(soup):
    spec=[];
    for ul in soup.select('div'):
        for li in ul.select('h3'):
            a=hasNumbers(li.text)
            if thing1 and a:
                spec.append(li.text)
    return spec

def parse10(soup):
    spec=[];
    for ul in soup.select('div'):
        for li in ul.select('h2'):
            a=hasNumbers(li.text)
            if thing1 and a:
                spec.append(li.text)
    return spec

def parse11(soup):
    spec=[];
    for ul in soup.select('div'):
        for li in ul.select('a'):
            a=hasNumbers(li.text)
            if thing1 and a:
                spec.append(li.text)
    return spec

def parse12(soup):
    spec=[];
    for ul in soup.select('table'):
        for li in ul.select('td'):
            temp=li.text
            a=hasNumbers(temp)
            if thing1 and a:
                spec.append(li.text)  
    return spec
def parse13(soup):
    spec=[];
    for ul in soup.select('ul'):
        for li in ul.select('li'):
            spec.append(li.text)         
    return spec

def parse14(soup):
    spec=[];
    for ul in soup.select('h4'):
        for li in ul.select('a'):
            temp=li.text
            a=hasNumbers(temp)
            if thing1 and a:
                spec.append(li.text)      
    return spec

def parse15(soup):
    spec=[];
    for ul in soup.select('tr'):
        for li in ul.select('td'):
            temp=li.text
            a=hasNumbers(temp)
            if thing1 and a:
                spec.append(li.text)      
    return spec

def parse16(soup):
    spec=[];
    for ul in soup.select('dl'):
        for li in ul.select('dt'):
            temp=li.text
            a=hasNumbers(temp)
            if thing1 and a:
                spec.append(li.text)      
    return spec


def parse17(soup):
    spec=[];
    for ul in soup.select('ol'):
        for li in ul.select('li'):
            temp=li.text
            a=hasNumbers(temp)
            if thing1 and a:
                spec.append(li.text)      
    return spec

def parse18(soup):
    spec=[];
    for ul in soup.select('div'):
        for li in ul.select('p'):
            spec.append(li.text) 
    return spec

def parse19(soup):
    spec=[];
    for ul in soup.select('h3'):
        spec.append(ul.text)      
    return spec

def parse20(soup):
    spec=[];
    for ul in soup.select('td'):
        for li in ul.select('a'):
            spec.append(li.text) 
    return spec
```
